```python
# whether there are no duplicates according to the unique number of the respondent.
dublicat_id = df_basic.duplicated('ResponseId').sum()
print(f'\n Number of duplicates = {dublicat_id}')
```

```python
ResponseId_count=df_basic.shape[0]
print(f"\n {ResponseId_count}")
```

```python
# We create a set of questions and immediately make an intersection
questions = set(df_structure['qname'].unique()) & set(df_basic.columns)
print(f"\n Common elements found: {len(questions)}")
```

```python
# We convert the set into a list for Pandas indexing
questions_list = list(questions)

# Add ResponseId for identification
cols_to_keep = ['ResponseId'] + questions_list

# We create the final working dataframe
df_final = df_basic[cols_to_keep]

print(f"\n New table dimension: {df_final.shape}")
```

```python
# We remove only those who did not answer ANY of these questions
df_cleaned = df_final.dropna(subset=questions_list, how='all')
print(f"\ n size of the cleared table: {df_cleaned.shape}")
```

```python
# Count the number of respondents without missing values
df_without_missing = df_cleaned[questions_list].notna().all(axis=1)
respondents_no_nan = df_without_missing.sum()
print(f"\n Full answers:{respondents_no_nan}")
```

```python
# Calculation of measures of central tendency
mean_exp = df_cleaned['WorkExp'].mean().round(1)
median_exp = df_cleaned['WorkExp'].median()
mode_exp = df_cleaned['WorkExp'].mode()[0]  # .mode() returns Series, we take the first element
# Outputting the result
print("\n Measures of Central Tendency for Work Experience (WorkExp):\n")
print(f" Mean    {mean_exp} ")
print(f" Median  {median_exp}")
print(f" Mode   {mode_exp}")
```

```python
# We filter respondents who chose "Remote"
remote_workers = df_cleaned[df_cleaned['RemoteWork'] == 'Remote']
# We count their number
remote_count = remote_workers.shape[0]
# We derive the final result
print(f"\n The number of respondents who work remotely: {remote_count}")
```

```python
# We are looking for the name of the column that contains the word 'Language'
matching_cols = [col for col in df_basic.columns if 'Language' in col]
print("Found similar columns:", matching_cols)
```

```python
# Define the name of the column
lang_col = 'LanguageHaveWorkedWith'

# We filter those who specified Python. we use .str.contains, na=False ignores empty values ​​(NaN)
python_users = df_basic[df_basic[lang_col].str.contains('python', case=False, na=False)]

#  Counting the number of Python users
python_count = python_users.shape[0]

# We count the total number of respondents (for the denominator)
total_respondents = df_basic.shape[0]
# We calculate the percentage
python_percentage = (python_count / total_respondents) * 100

# We derive the result
print(f"\n Percentage of respondents who program in Python: {python_percentage:.1f}%")
```

```python
# We are looking for the name of the column that contains the word 'learn'
matching_cols = [col for col in df_basic.columns if 'Learn' in col]
print("\n Found similar columns:", matching_cols)
```

```python
# Define the name of the column
learn_col = 'LearnCode'

# We filter those who have chosen online courses, we use 'Online Courses'
online_learners = df_basic[df_basic[learn_col].str.contains('Online Courses', case=False, na=False)]

# We count the number
online_count = online_learners.shape[0]

# We derive the result
print(f"\n The number of respondents who studied through online courses: {online_count}")
```

```python
# Looking for a column name that contains the word 'Country'
matching_cols = [col for col in df_basic.columns if 'Country' in col]
print("\n Знайдені схожі колонки:", matching_cols)
```

```python
# We will use the ready-made dataset on Python developers from the previous task 5 python_users
# the number of Python users from the previous task
print (f"\n Number of Python users: {python_count}")

df_python = python_users.dropna(subset=['ConvertedCompYearly'])
python_clean = df_python.shape[0]
print (f"\n Number of Python users who reported compensation: {python_clean}")

# We group by countries and count metrics
# Add 'count' to see the size of the sample for each country
country_analysis = df_python.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median', 'count'])

# Rename for convenience
country_analysis.columns = ['mean ($)', 'median ($)', 'Number of respondents']

# We filter countries with at least 5 developers for statistical reliability
final_table = country_analysis[country_analysis['Number of respondents'] > 5].sort_values(by='median ($)', ascending=False)

# We round and change the type to an integer for beauty
final_table = final_table.round(0).astype(int)

# Output of the result
print("\n Salary analysis of Python developers by country:")
display(final_table.head(15))
```

```python
# Since df_python no longer has NaNs in salaries, we just sort it
# Sort by decreasing compensation
df_top_paid = df_python.sort_values(by='ConvertedCompYearly', ascending=False)

# We select the TOP-5 respondents and the required columns
# we add 'Country' out of curiosity
top_5 = df_top_paid[['ConvertedCompYearly', 'EdLevel', 'Country']].head(5)

# We derive the result
print("\n Education level of the 5 highest paid Python developers:")
display(top_5)
```

```python
# We count the total number of respondents in each age category
age_total = df_basic['Age'].value_counts()

# We count the number of Python developers in each age category
# We count the number of Python developers in each age category
age_python = python_users['Age'].value_counts()

# We combine these data into one table
age_categoric = pd.concat([age_total, age_python], axis=1)
age_categoric.columns = ['Total', 'Python_Users']

# We calculate the percentage
age_categoric['(%)'] = (age_categoric['Python_Users'] / age_categoric['Total']) * 100

# We round the result for a better look
age_categoric['(%)'] = age_categoric['(%)'].round(1)

# Sort by percentage to see where Python is most popular
# Or you can sort by index to see ages one by one
age_categoric = age_categoric.sort_values(by='(%)', ascending=False)

# We derive the result
print("\n Popularity of Python among different age groups:")
display(age_categoric[['Total','(%)']])
```

```python
# Searching for the column name that contains the word 'Ind'
matching_cols = [col for col in df_basic.columns if 'Ind' in col]
print("\n Found similar columns:", matching_cols)
```

```python
# We filter remote workers
remote_respond = df_basic[df_basic['RemoteWork'] == 'Remote']

# We find the 75th quantile (the limit above which the top 25% of salaries are located)
q75 = remote_respond['ConvertedCompYearly'].quantile(0.75)
print(f"\n Limit of the 75th quantile (Top-25%): ${q75:,.0f}")

# We choose only those who earn the same or more
top_remote = remote_respond[remote_respond['ConvertedCompYearly'] >= q75]

# We count the number of respondents by industry.
# The value_counts() method is more convenient here, because it immediately sorts from largest to smallest
industry_counts = top_remote['Industry'].value_counts()

# Convert to DataFrame for good output
industry_report = industry_counts.reset_index()
industry_report.columns = ['Industry', 'Number of top specialists']

print("\n The most popular industries among high-paying remote workers:")
display(industry_report.head(10))
```
